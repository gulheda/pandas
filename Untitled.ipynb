```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline  
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors = 'coerce')
df['year_added'] = df['date_added'].dt.year
df['duration_minutes'] = df['duration'].str.extract('(\d+)').astype(float)
```

```python
df['year_added'].value_counts().sort_index().plot(kind='line', figsize=(10,5)) #çizgi grafiği oluştur
plt.title("yıllara göre netflix'e eklelnen içerik sayısı")
plt.xlabel("yıl")
plt.ylabel("içerik sayısı")
plt.grid(True)
plt.show()
```

```python
def en_cok_icerik_ureten_ulkeleri_goster(veri: pd.DataFrame, n: int = 10) -> None:
    """
    En çok içerik üreten N ülkeyi gösteren bir çubuk grafik oluşturur.
    
    Args:
        veri (pd.DataFrame): 'country' sütunu içeren DataFrame
        n (int, optional): Gösterilecek ülke sayısı. Varsayılan değer 10.
    
    Returns:
        None: Grafiği ekranda gösterir
    """
    # En çok içerik üreten N ülkeyi al
    top_ulkeler = veri['country'].value_counts().head(n)
    
    # Grafiği oluştur
    plt.figure(figsize=(10, 5))
    top_ulkeler.plot(kind='bar')
    
    # Etiketleri ve başlığı ekle
    plt.title("En çok içerik üreten {} ülke".format(n))
    plt.xlabel("Ülke")
    plt.ylabel("İçerik sayısı")
    plt.xticks(rotation=45)
    plt.tight_layout()  # Ülke isimlerinin tam görünmesini sağlar
    
    # Grafiği göster
    plt.show()

# Kullanım örneği:
en_cok_icerik_ureten_ulkeleri_goster(df)
```
